# Inference Engine & Agent Integration
This notebook demonstrates how downstream services (e.g., LangGraph agents or Backend APIs) will interact with the ML pipeline to score transactions in real-time.

```python
import sys
import os
import json
sys.path.append(os.path.abspath('../../'))

from ml.inference.risk_model import RiskInferenceEngine

# Initialize Engine (loads model from registry natively)
engine = RiskInferenceEngine()
```

```python
# Define a mock real-time transaction payload
# In production, the backend joins the customer profile and recipient info into this flat structure.
payload = {
    "transaction_id": "DEMO-TXN-999",
    "amount": 85000,
    "average_transaction": 5000,
    "current_balance": 90000,
    "timestamp": "2026-07-02T15:00:00",
    "trusted_recipient": False,
    "transaction_count": 0,
    "trust_score": 20,
    "reported_scam_count_rep": 2,
    "favorite_categories": "Shopping,Food",
    "category": "Transfer",
    "usual_location": "Mumbai",
    "city": "Delhi",
    "usual_device": "Mobile",
    "device_type_x": "Web",
    "risk_profile": "Low",
    "digital_payment_ratio": 0.9,
    "transaction_type": "Debit",
    "payment_channel": "RTGS"
}
```

```python
# Execute Prediction
response = engine.predict(payload)

print(json.dumps(response, indent=2))
```

### Agent Trigger Logic
The output JSON above is directly parsed by the Agentic Workflow:
- **If `risk_level` is Medium**: Agent surfaces an educational warning in the UI.
- **If `risk_level` is High**: Agent blocks transaction and initiates an interactive conversation, asking for the contextual purpose of the transaction.
- **If `risk_level` is Critical**: Agent blocks transaction and heavily recommends cancellation, escalating to a human fraud specialist if the user insists.
