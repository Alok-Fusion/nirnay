# NIRNAY - Exploratory Data Analysis
This notebook provides an interactive environment to explore the synthetic datasets, visualize feature distributions, and analyze the engineered ML features.

```python
import sys
import os
# Add the project root to the python path
sys.path.append(os.path.abspath('../../'))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ml.data.joiner import DataJoiner
from ml.feature_engineering.pipeline import FeatureEngineer

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
# 1. Load and Join Datasets
joiner = DataJoiner()
df = joiner.create_master_dataset()
print(f'Master dataset shape: {df.shape}')
df.head()
```

```python
# 2. Analyze Target Variable Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='risk_label', palette='Set2')
plt.title('Distribution of Safe (0) vs Unsafe (1) Transactions')
plt.show()

print(df['risk_label'].value_counts(normalize=True))
```

```python
# 3. Feature Engineering
fe = FeatureEngineer()
df_features = fe.fit_transform(df)
print(f'Engineered dataset shape: {df_features.shape}')
df_features.head()
```

```python
# 4. Visualize Key Engineered Features
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Amount Deviation by Risk
sns.boxplot(data=df_features, x='risk_label', y='amount_deviation', ax=axes[0])
axes[0].set_title('Amount Deviation vs Risk Label')
axes[0].set_ylim(0, 20) # Cap y-axis for better visibility

# Balance Ratio by Risk
sns.boxplot(data=df_features, x='risk_label', y='balance_ratio', ax=axes[1])
axes[1].set_title('Balance Ratio vs Risk Label')

plt.tight_layout()
plt.show()
```
